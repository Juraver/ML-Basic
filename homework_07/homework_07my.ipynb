```python
z = np.arange(10,50)
print('---Матрица Z--------')
print(z[::-1])
print('--------------------')
def chet(y):
 
  return y[y % 2 == 0]

print('---Четные элементы из Матрицы Z------')
print(chet(z))
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
x = np.random.rand(7, 7)
print('--------Матрица Х--------')
print(x)
print('-------------------------')
print(f'Среднее = {np.mean(x)}')
print(f'Стандартное = {np.std(x)}')
print('-------------------------')
def normalize(x):
    return (x - x.mean(axis=0)) / x.std(axis=0)
print('--------Матрица Нормализации--------')
print(normalize(x))
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import random
import seaborn as sns
df = pd.read_csv('tips.csv', sep=',')

bins = int(1 + np.log2(244))

plt.hist(df['total_bill'], bins=bins, color='g', edgecolor='black')

plt.xlabel('total_bill')
plt.ylabel('Частота')
plt.title('Гистограмма распределения total_bill')

plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
df.plot.scatter(x='total_bill', y='tip')
```

### Постройте pairplot

```python
sns.pairplot(df)
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
df['day'].replace('Mon', 1, inplace=True)
df['day'].replace('Tue', 2, inplace=True)
df['day'].replace('Wed', 3, inplace=True)
df['day'].replace('Thur', 4, inplace=True)
df['day'].replace('Fri', 5, inplace=True)
df['day'].replace('Sat', 6, inplace=True)
df['day'].replace('Sun', 7, inplace=True)

df.plot.scatter(x='day', y='total_bill')
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
df_dinner = df[df['time'] == 'Dinner']

bins = int(1 + np.log2(122))

plt.hist(df_dinner['total_bill'], bins=bins, color='g', edgecolor='black')

plt.xlabel('total_bill on Dinner')
plt.ylabel('Частота')
plt.title('Гистограмма распределения total_bill on Dinner')

plt.show()

df_lunch = df[df['time'] == 'Lunch']

bins = int(1 + np.log2(122))

plt.hist(df_lunch['total_bill'], bins=bins, color='g', edgecolor='black')

plt.xlabel('total_bill on Lunch')
plt.ylabel('Частота')
plt.title('Гистограмма распределения total_bill on Lunch')

plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
_,ax = plt.subplots(1, 2, figsize=(16, 9))

df_male = df[df['sex'] == 'Male']
sns.scatterplot(ax = ax[0], data = df_male, x='tip', y='total_bill', hue = 'smoker', s = 50, alpha = 0.5)

df_female = df[df['sex'] == 'Female']
sns.scatterplot(ax = ax[1], data = df_female, x='tip', y='total_bill', hue = 'smoker', s = 50, alpha = 0.5)
plt.title('График взаимосвязи между признаками total_bill и tip для Male(слева) и для Fеmale(справа) в зависимости от smoker')
```
